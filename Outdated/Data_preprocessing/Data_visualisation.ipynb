```python
import pandas as pd

file_name = "Masters"
df = pd.read_csv(f'../../data/{file_name}_CleanedUp.csv')

print(df.columns)
```

```python
# Plot the images of the first 10 rows
import os
from PIL import Image
import matplotlib.pyplot as plt
from time import sleep

img_lists = os.listdir(f"../../images/{file_name}/")

img_count = 100
for i in range(img_count):
    # Get the ID of the post
    post_id = str(df.iloc[i]['Id'])
    imgs = [img for img in img_lists if post_id in img]  
    for j, img in enumerate(imgs):
        plt.figure(figsize=(6, 6))
        img_path = f"../../images/{file_name}/{img}"
        img = Image.open(img_path)
        plt.imshow(img)
        plt.axis('off')
        print(f"Post: {post_id} - {j}")
        print(f"Alt text: {df.iloc[i]['AltText']}")
        print(f"Caption: {df.iloc[i]['NewBody']}")
        plt.show()
        sleep(.1)
```
