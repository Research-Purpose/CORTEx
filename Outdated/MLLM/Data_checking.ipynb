```python
import pandas as pd
import csv

# Define the path to your CSV file
file_path = '../../data/Results_2.csv'
corrected_file_path = '../../data/Results_corrected_2.csv'

# Define the expected number of fields
expected_fields = 6

# Open the input and output files
with open(file_path, 'r', newline='', encoding='utf-8') as infile, \
     open(corrected_file_path, 'w', newline='', encoding='utf-8') as outfile:
    
    reader = csv.reader(infile, delimiter=',', quotechar='"')
    writer = csv.writer(outfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    for line in reader:
        # Check if the line has the expected number of fields
        if len(line) == expected_fields:
            writer.writerow(line)
        # else:
            # print(f"Warning: Line has {len(line)} fields (expected {expected_fields}): {line}")
```

```python
# Load the corrected CSV file into a DataFrame
df = pd.read_csv(corrected_file_path, quotechar='"')

# Set the column names
df.columns = ['ID', 'ImgName', 'Question', 'Text1', 'Text2', 'ParagraphID']
```

```python
# Show the amount of None values
none_count = df['ParagraphID'].isnull().sum()
total_count = len(df)

print(f"Percentage of None values: {none_count / total_count * 100:.2f}%")
```

This shows that 49% of the citations could not be found in the images, this could be because the image just doesn't include text, or because of hallucination

```python
# Iterate through the rows of the corrected CSV file
# Show the image using matplotlib

from matplotlib import pyplot as plt
import cv2

img_dir = '../../images/Output_2/RESULT/'

# Randomly shuffle the DataFrame 
df = df.sample(frac=1).reset_index(drop=True)

for index, row in df.iterrows():
    # Print the Question, Text1, Text2, CitationType, and ParagraphID
    print(f"ID: {row['ID']}")
    print(f"Question: {row['Question']}")
    print(f"Text1: {row['Text1']}")
    print(f"Text2: {row['Text2']}")

    img_name = row['ImgName']
    img_path = img_dir + img_name
    img = cv2.imread(img_path)
    try:
        plt.imshow(img)
        plt.axis('off')
        plt.show()
    except:
        pass
```
